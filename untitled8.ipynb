```python
# CHECK POINT DE FONTIONS PYTHON
# Voici une fonction calculateur qui effectue les quatre opérations de base :
def calculateur(num1, num2):
    """
    Effectue les quatre opérations arithmétiques de base sur deux nombres.

    Paramètres:
    num1 (float/int): Premier nombre
    num2 (float/int): Deuxième nombre

    Retourne:
    tuple: (addition, soustraction, multiplication, division)
    """
    # Addition
    addition = num1 + num2

    # Soustraction
    soustraction = num1 - num2

    # Multiplication
    multiplication = num1 * num2

    # Division avec vérification pour éviter la division par zéro
    if num2 != 0:
        division = num1 / num2
    else:
        division = "Erreur: Division par zéro"

    return addition, soustraction, multiplication, division

   #Résolution du Calculateur
if __name__ == "__main__":
    # Test avec deux nombres
    resultats = calculateur(10, 5)
    print(f"Addition: {resultats[0]}")
    print(f"Soustraction: {resultats[1]}")
    print(f"Multiplication: {resultats[2]}")
    print(f"Division: {resultats[3]}")

    print("\n--- Autre test avec des nombres différents ---")
    a = 8
    b = 2
    add, sous, mult, div = calculateur(a, b)
    print(f"{a} + {b} = {add}")
    print(f"{a} - {b} = {sous}")
    print(f"{a} × {b} = {mult}")
    print(f"{a} ÷ {b} = {div}")
```
