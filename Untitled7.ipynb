1. Utilisez la fonction « input » pour inviter le joueur à saisir sa proposition sous forme de chaîne de caractères.
2. Utilisez la fonction 'int' pour convertir la proposition du joueur en un entier.
3. Utilisez l'instruction 'import' pour importer le module 'random'.
4. Utilisez la fonction 'random.randint()' pour générer un entier aléatoire entre 1 et 100.
5. Utilisez une instruction « if/else » pour vérifier si la proposition du joueur est trop élevée, trop basse ou correcte.
6. Utilisez la fonction « print » pour fournir un retour d’information au joueur en fonction de sa supposition.
7. Utilisez une boucle « while » pour continuer à demander des devinettes au joueur jusqu'à ce qu'il trouve le bon nombre.

```python
# Utilisation de la fonction « input » pour inviter le joueur à saisir sa proposition sous forme de chaîne de caractères
import random

ger = random.randint(1,100)
devinette = 0
while devinette != ger :
  devinette = int(input('Donner un nombre: '))
  if devinette > ger:
    print('trop grand')
  elif devinette < ger:
    print('trop petit')
  else:
    print('Félicitation')
```

```python
# Utilisation de la fonction 'int' pour convertir la proposition du joueur en un entier.
steps = 0

n = int(input("Entrez un nombre : "))

while n != 1:
    if n % 2 == 0:
        n = n // 2
    else:
        n = 3*n + 1
    print("→", n, end=" ")
    steps += 1

print("\nNombre d'étapes :", steps)
```
